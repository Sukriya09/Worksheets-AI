 What percent of the total number of passengers on the Titanic belonged to each age group
 (e.g., child, adult, senior)?
 To answer the question perform following operation
  Create a new column, age
 group, that categorizes passengers into "child" (age < 18), "adult" (age
 18{64), and "senior" (age 65 and above)

```python
def cat_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(cat_age)
df.head()
```

 2. Calculate the total number of passengers on the Titanic.


```python
total_passengers = df['PassengerId'].count()
total_passengers
```

 3. Count the number of passengers in each age group.


```python
df['age_group'].value_counts()
```

 4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
 in each age group

```python
adult_proportion = df['age_group'].value_counts()['adult'] / total_passengers
print(adult_proportion*100)
```

```python
senior_proportion = df['age_group'].value_counts()['senior']/total_passengers
print(senior_proportion*100)
```

```python
child_proportion = df['age_group'].value_counts()['child']/total_passengers
print(child_proportion*100)
```
