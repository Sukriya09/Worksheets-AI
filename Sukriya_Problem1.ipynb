```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/bank.csv")
```

```python
df
```

Identify columns with dtypes = object

```python
object = df.select_dtypes(include='object').columns
print(object)#provide column name
```

*unique* values of those columns

```python
for col in object:
    unique_values = df[col].unique()
    print(f"Unique values in column '{col}':")
    print(unique_values)
    print()
```

check for the total number of null values in each column

```python
null_values = df.isnull().sum()
print(null_values)
```

Drop all the colimns with dtypes object and store in new DataFrame, also write the DataFrame in ".csv" withh name "banknumbericdara.csv"

```python
numeric_df = df.drop(columns = object)
numeric_df.to_csv("banknumericdata.csv", index = False)

numeric_data = pd.read_csv("banknumericdata.csv")
summary = numeric_data.describe()
print(summary)
```
